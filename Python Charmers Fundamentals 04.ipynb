```python
import pandas as pd

taxis = pd.read_csv(r"C:\Users\TylerHa\Documents\GitHub\Python\Data\2019_Yellow_Taxi_Trip_Data.csv")
taxis.head()
```

```python
columns_to_drop = ['vendorid','ratecodeid','store_and_fwd_flag','pulocationid','dolocationid']
taxis = taxis.drop(columns=columns_to_drop)
taxis.head()
```

```python
taxis = taxis.rename(
    columns={
        'tpep_pickup_datetime': 'pickup',
        'tpep_dropoff_datetime': 'dropoff'
    }
)
taxis.columns
```

```python
taxis.head()
```

```python
taxis.dtypes
```

```python
taxis[['pickup','dropoff']] = taxis[['pickup','dropoff']].apply(pd.to_datetime)
taxis.dtypes
```

```python
taxis['new_column1'] = 'This is a new column'
taxis = taxis.assign(new_column2 = lambda x: 'This is a new column too')
taxis.head(2)
```

```python
taxis['trip_time1'] = taxis['dropoff'] - taxis['pickup']
taxis = taxis.assign(trip_time2 = lambda x: x.dropoff - x.pickup)
taxis.head(2)
```

```python
more_columns_todrop = ['new_column1','new_column2','trip_time1','trip_time2']
taxis = taxis.drop(columns=more_columns_todrop)
taxis.head()
```

```python
taxis['elapsed_time'] = taxis['dropoff'] - taxis['pickup']
taxis['cost_beforetip'] = taxis['total_amount'] - taxis['tip_amount']
taxis['tip_percentage'] = taxis['tip_amount'] / taxis['cost_beforetip']
taxis['total_taxestollsfees'] = taxis['cost_beforetip'] - taxis['fare_amount']
taxis['total_seconds'] = taxis.elapsed_time.dt.total_seconds()
taxis
```

```python
onemore_column_todrop = ['avg_speed',]
taxis = taxis.drop(columns=onemore_column_todrop)
taxis
```

```python
taxis['avg_speed'] = taxis['trip_distance']/taxis['total_seconds']*60*60
taxis
```
