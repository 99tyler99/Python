```python
import pandas as pd

tsa = pd.read_csv(r"C:\Users\TylerHa\Documents\GitHub\Python\Data\tsa_passenger_throughput.csv", parse_dates=['Date'])
tsa.head()
```

```python
tsa = tsa.rename(columns=lambda x: x.lower().split()[0])
tsa.head()
```

```python
tsa_melted = pd.melt(tsa, 
    id_vars=['date'],
    var_name='year',
    value_name='travelers'
    )
tsa_melted.sample(5, random_state=1)
```

```python
tsa_melted['date']=pd.to_datetime(tsa_melted['year'] + tsa_melted['date'].dt.strftime('-%m-%d'))
tsa_melted.sample(5, random_state=1)
```

```python
tsa_melted.sort_values('date')
```

```python
tsa_melted = tsa_melted.dropna()
tsa_melted.sort_values('date')
```

```python
tsa_melted['date']=pd.to_datetime(tsa_melted['date'])

first_10_days = tsa_melted.loc[(tsa_melted['date'].dt.month == 3)&(tsa_melted['date'].dt.day <= 10)].copy()
first_10_days['day_in_march'] = first_10_days['date'].dt.day

first_10_days_pivot = pd.pivot(first_10_days,index='year', columns='day_in_march', values='travelers')
first_10_days_pivot
```

```python
tsa_pivoted = first_10_days_pivot
tsa_pivoted
```

```python
tsa_pivoted=tsa_pivoted.transpose()
tsa_pivoted
```

```python
holidays = pd.read_csv(r"C:\Users\TylerHa\Documents\GitHub\Python\Data\holidays.csv", parse_dates=True, index_col='date')
holidays.loc['2019']
```

```python
tsa_melted_holidays = tsa_melted.merge(
    holidays,
    left_on='date',
    right_index=True,
    how='left'
)
tsa_melted_holidays = tsa_melted_holidays.sort_values('date')
tsa_melted_holidays
```

```python
tsa_melted_holiday_travel = tsa_melted_holidays.assign(
    holiday=lambda x:
        x.holiday.ffill(limit=1).bfill(limit=2)
)
tsa_melted_holiday_travel
```

```python
tsa_melted_holiday_travel.loc[
    (tsa_melted_holiday_travel['year']=='2019') &
    tsa_melted_holiday_travel['holiday'].notnull()
    ]
```

```python
pd.pivot_table(
    tsa_melted_holiday_travel,
    index = 'year',
    columns = 'holiday',
    values = 'travelers',
    aggfunc = 'sum'
)
```
